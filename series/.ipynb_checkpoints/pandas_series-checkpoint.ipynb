# Pandas -

Pandas is a powerful Python library used for data manipulation, <br>
cleaning, analysis, and exploration. <br>
It is built on top of NumPy and is widely used in Data Science, Data Analysis, and Machine Learning.

# Key Features -

1. Fast and efficient data structures
2. Handling missing data easily
3. Data alignment and reshaping
4. Powerful group-by and aggregation
5. Easy data import/export (CSV, Excel, SQL, JSON)

# Series form list -

```python
country = ["India","USA","China","Nepal","Pakistan"]
pd.Series(country)
```

```python
# Interger : -
run =[15,20,30,45,65,84]
pd.Series(run)
```

```python
# Custom Index :
marks=[64,25,85,42]
subjects=["Maths","Science","Hindi","English"]
pd.Series(marks,index=subjects)
```

```python
# setting the name :
pd.Series(marks,index=subjects,name="Shubham Marks")
```
